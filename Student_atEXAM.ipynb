```python
#Let's findout students who taken test preparation course
(data['TEST PREPARATION COURSE'].value_counts())
```

```python
(data['TEST PREPARATION COURSE'].value_counts()/len(data))*100
```

```python
#Donut chart
testcourse = data['TEST PREPARATION COURSE'].value_counts()
labels = data["TEST PREPARATION COURSE"].unique()
plt.pie(testcourse,labels=labels,autopct="%1.1f%%",
        shadow=True,explode=(0.025,0.025),
        pctdistance=0.85,startangle=90)
plt.gcf().gca().add_artist(plt.Circle((0,0),0.70,fc='blue'))
plt.title('TEST PREPARATION COURSE',fontsize=12)
plt.tight_layout()
plt.show()
```

As here nearly two third didn't took preparation course.

```python
testpc = data.groupby("TEST PREPARATION COURSE")
testpc.mean()
```

So here the outcomes only one third around students take preparation course but thier scores are improved.

```python
#let's visulize how preparation helped student at genderwise
##Here I will use subplots to plot different plots in one diagram to compare easily

fig, ax = plt.subplots(1, 3, figsize=(14,7))
sns.violinplot(x="TEST PREPARATION COURSE", y='MATH SCORE', 
               data=data,hue='GENDER',split=True,palette='PuRd',
               ax = ax[0])
sns.violinplot(x="TEST PREPARATION COURSE", y='READING SCORE', 
               data=data,hue='GENDER',split = True, 
               palette='Purples', 
               ax = ax[1])
sns.violinplot(x="TEST PREPARATION COURSE", y='WRITING SCORE', 
               data=data,hue='GENDER',split = True, 
               palette='RdPu', 
               ax = ax[2])
```

```python
## Boxplot to interper avg with preparation
sns.boxplot(x="TEST PREPARATION COURSE", y="AVERAGE", hue = "GENDER", data = data)
```

As here we notice test prep course helps to improve thier overall score.

```python
#Let's find out the percentage of students who receive Standard and Reduced Lunch
(data['LUNCH'].value_counts()/len(data)) * 100
```

```python
# Plotting the figures with swarmplot for ethnicity, subject and luch
fig, ax = plt.subplots(3, 1, figsize=(16,16))
sns.swarmplot(x="RACE/ETHNICITY", y='MATH SCORE', 
              data=data,hue='LUNCH',palette='Purples', ax = ax[0])
sns.swarmplot(x="RACE/ETHNICITY", y='READING SCORE', 
              data=data,hue='LUNCH', palette='Blues', ax = ax[1])
sns.swarmplot(x="RACE/ETHNICITY", y='WRITING SCORE', 
              data=data,hue='LUNCH', palette='Greens', ax = ax[2])
```
